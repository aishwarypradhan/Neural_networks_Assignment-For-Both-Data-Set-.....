```python
model_data = df[['CDP', 'GTEP','TIT', 'TAT', 'AFDP', 'CO', 'AT',"TEY"]]
model_data.head()
```

### 5.2 Feature Engineering
#### Applying some Data Transformation to increase the linear realtionship and improve our model prediction as well it scores
### Applying Standard Scaler

```python
continuous_feature=[feature for feature in model_data.columns if model_data[feature].dtype!='O']
print('Continuous Feature Count {}'.format(len(continuous_feature)))
```

```python
df_standard_scaled = model_data.copy()
features = df_standard_scaled[continuous_feature]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

df_standard_scaled[continuous_feature] = scaler.fit_transform(features.values)
df_standard_scaled
```

```python
print('Mean' '\n',np.round(df_standard_scaled.mean(),1))
```

```python
print('Standard Devaition','\n',np.round(df_standard_scaled.std(),1))
```

### 5.3 Test Train Split With Imbalanced Dataset

```python
x = df_standard_scaled.drop('TEY',axis=1)
y = df_standard_scaled[['TEY']]
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
model=Sequential()
model.add(Dense(12,input_dim=7,activation='sigmoid'))###Input Layer
model.add(Dense(18,input_dim=12,activation='relu'))
model.add(Dense(24,input_dim=18,activation='sigmoid'))
model.add(Dense(1,activation="sigmoid"))### Out put Layer
```

```python
model.compile(loss= 'mean_absolute_error', optimizer= 'adam', metrics= ["mean_absolute_error"])
```

```python
from pandas.core.common import random_state
history=model.fit(x,y,validation_split=0.20,epochs=200,batch_size=50)
```

```python
scores=model.evaluate(x,y)
print((model.metrics_names,scores))
```
